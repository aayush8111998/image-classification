```python
import matplotlib.pyplot as plt 
import numpy as np 
import tensorflow as tf
from tensorflow.keras import models, datasets, layers
from keras.preprocessing import image
from PIL import Image

#load dataset
(xt1,yt1), (xt2,yt2)= datasets.cifar10.load_data()
 
#define class
classes= ["aeroplane", "automobile", "bird","cat", "deer", "dog", "frog", "horse", "ship", "truck"]

#resize image
xt1= xt1/255.0
xt2= xt2/255.0

#cnn model(2 layers)
cnn= tf.keras.models.Sequential()
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
#flattening
cnn.add(tf.keras.layers.Flatten())
#full connection
cnn.add(tf.keras.layers.Dense(64, activation='relu'))

#o/p

cnn.add(tf.keras.layers.Dense(10,activation='softmax'))

#compile cnn
cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

#train cnn
cnn.fit(xt1, yt1, epochs= 25)

test_img=image.load_img(r"D:\datasets\Section+40+-+Convolutional+Neural+Networks+(CNN)\Section 40 - Convolutional Neural Networks (CNN)\dataset\single_prediction\house.jpg", target_size=(32,32))
test_img=image.img_to_array(test_img)
test_img= np.expand_dims(test_img, axis=0)
result= cnn.predict(test_img)

#print(test_img)
y_pred= cnn.predict(xt2)
#y_pred[:5]



y_class=[np.argmax(element) for element in y_pred]
#y_class[:5]

#classes[y_class[6]]

#print(len(result))


#index=np.where(result=='1')
#print(index)
#print(result)

#print(result[0][7])
#print(result[0][8])




result= result.astype(np.int32)
#print(result.dtype)

im=Image.open("D:\datasets\Section+40+-+Convolutional+Neural+Networks+(CNN)\Section 40 - Convolutional Neural Networks (CNN)\dataset\single_prediction\house.jpg")
plt.imshow(im)


if result[0][0]==1:
    print('aeroplane')
if result[0][1]==1:
    print('automobile')
if result[0][2]==1:
    print('bird')
if result[0][3]==1:
    print('cat')
if result[0][4]==1:
    print('deer')
if result[0][5]==1:
    print('dog')
if result[0][6]==1:
    print('frog')
if result[0][7]==1:
    print('horse')
if result[0][8]==1:
    print('ship')
if result[0][9]==1:
    print('truck')
    

```
